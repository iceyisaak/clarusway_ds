```python
# Convert all df['mass'] into float
# df['mass'] = pd.to_numeric(df['mass'], errors='coerce')


# Alternative way??????????????
df['mass'] = df['mass'].str.replace(',','')
df['mass'] = df['mass'].astype('float')

# See converted data
# df['mass'].sort_values(ascending=False)
df['mass']
```

```python
# Assign categories to data

# High_mass = df.loc[df['mass'] >= 100, ['mass']]
# Normal_mass = df.loc[(df['mass'] > 50) & (df['mass'] < 100),['mass']]
# Low_mass = df.loc[df['mass'] <= 50,['mass']]
```

```python
# Function to categorise mass
def categorise_mass(x):
    if x >= 100:
        return 'High Mass'
    elif x > 50 and x < 100:
        return 'Normal Mass'
    elif x <= 50:
        return 'Low Mass'
    else:
        return np.nan
```

```python
# Apply categorise_mass for 'mass' col + put result in 'mass_cat' col
df["mass_cat"] = df['mass'].apply(categorise_mass)

# Count values in 'mass_cat'
mass_cat_chart = df['mass_cat'].value_counts(dropna=False)

mass_cat_chart
```

```python
# with pandas

plt.figure(figsize=(10,8))
mass_cat_chart.plot.pie(autopct=lambda x: f'{round(x,2)}%')
# mass_cat_chart.plot.bar();

plt.ylabel('')
plt.legend();
```

```python
# mass_cat_chart.index
```

```python
# with matplotlib

plt.figure(figsize=(10,8))
plt.title('Starwars Characters by Mass', fontsize=18, color='#101010')

chart_statfig = mass_cat_chart.values
chart_labels = mass_cat_chart.index

plt.pie(chart_statfig, labels=chart_labels, autopct= lambda x: f'{round(x,2)}%')
plt.legend();
```

#### **Note: Calculate 1 BBY as 1 Year**
1 BBY = 1 Year


```python
#Remove text in birth_year

#Select 'birth_year'
birth_year = df.loc[:,'birth_year']

#Replace text with blank
df['birth_year'] = birth_year.str.replace(r'\D+','',regex=True)

df['birth_year']
```

```python
#Convert birth_year to float
df['birth_year'] = df['birth_year'].astype(float)
# df['birth_year'] = pd.to_numeric(df['birth_year'], errors='coerce')

df['birth_year']
```

```python
df.info()
```

```python
# Get name with birth year > 100
characters_mt_100yo = df.loc[df['birth_year'] > 100,['name','birth_year']]

characters_mt_100yo
```

```python
character_name_by_yr = characters_mt_100yo.groupby('name')['birth_year'].sum().sort_values()
# character_name_by_yr = characters_mt_100yo.sort_values(by='birth_year')

character_name_by_yr
```

```python
# with pandas

plt.figure(figsize=(12,8))

character_name_by_yr.plot.barh()
age_chart = plt.barh(character_name_by_yr.index,character_name_by_yr.values)
plt.title('Oldest Starwars Characters by Age', fontsize=18, color='#101010')
plt.xlabel('Years')
plt.ylabel('Name')
plt.bar_label(age_chart)

plt.show()
```

```python
character_name_by_yr
```

```python
# with matplotlib
plt.figure(figsize=(12,8))

plt.title('Oldest Starwars Characters by Age', fontsize=18, color='#101010')
plt.xlabel('Years')
plt.ylabel('Name')

age_chart = plt.barh(character_name_by_yr.index,character_name_by_yr.values)
plt.bar_label(age_chart);
```
